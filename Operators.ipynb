```python
10 + 3
# int + int -> int
```

```python
10 - 3
# int - int -> int
```

```python
10 * 3
# int * int -> int
```

```python
10 / 3
# int / int -> float
# / is a special operator
# It always gives float value as answer
```

```python
5.0 + 5.0 
# float + float -> float
# float - float -> float
# float * float -> float
# float / float -> float
```

```python
3 + 5.0

# integer (operator) float -> float
# float (operator) integer -> float
```

```python
x = 10
y = 2.5
print(x / y)

print(3.5+3.5)

print(1 + 2.0)
```

```python
# Power operator

print(3**3)
print(2**1)
print(3.7**2)
print(3.7**1.2)
```

```python
# Floor division //

print(10 // 3)
print(-10 // 3)
print(2 // 2.0)
print(2 // 4.0)
# floor(2 / 4.0) -> floor(0.5) -> 0.0
```
